```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
total_revenue=df["SALES"].sum()/1000000
total_revenue=total_revenue.round(2)
print(f"Total Revenue: {total_revenue} millions")
```

```python
df.groupby("TERRITORY")["SALES"]
```

```python
total_sales = df.groupby("TERRITORY")
print(total_sales)
```

```python
total_sales = df.groupby("TERRITORY")["SALES"].sum().sort_values(ascending=False)
print(total_sales)
```

```python
df_total_sales = total_sales.reset_index()
df_total_sales.columns = ["Region","Amount"]
```

```python
df_total_sales
```

```python
total_sales_by_country = df.groupby("COUNTRY")["ORDERNUMBER"].count().sort_values(ascending=False)
df_total_sales_by_country = total_sales_by_country.reset_index()
df_total_sales_by_country.columns = {"Country","No_Of_Order"}
```

```python
df_total_sales_by_country
```

```python
most_bought_by_country=df.groupby("COUNTRY")["PRODUCTLINE"].agg(lambda x: x.value_counts().idxmax())
```

```python
df_most_bought_by_country=most_bought_by_country.reset_index()
df_most_bought_by_country.columns = {"Country","Most Frequently Bought"}
```

```python
df_most_bought_by_country
```
